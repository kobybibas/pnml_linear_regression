```python
# Take the mean over experimetns
res_list = []
for i, res_df in enumerate(res_dfs):

    # Verify each df is for specific dataset
    dataset_name = res_df.dataset_name.unique()
    assert len(dataset_name) == 1
    dataset_name = dataset_name[0]
    print("Dataset: {}. {}".format(i, dataset_name))

    # Filter samples that the pNML was not converged there from empirical reasons
    print("Failure: {}. Inf regret {} Inf logloss. {} empirical>analytical {}/{}".format(
            (res_df["pnml_success"] == False).sum(),
            (res_df["pnml_regret"] >= log_inf_limit).sum(),
            (res_df["pnml_test_logloss"] >= log_inf_limit).sum(),
            (res_df["pnml_regret"] > res_df["analytical_pnml_regret"] + 1e-1).sum(),
            len(res_df),
        )
    )
    res_df = res_df[res_df["pnml_success"] == True]
    
    
    
    
#     res_df = res_df[res_df["pnml_regret"] < log_inf_limit]
#     res_df = res_df[res_df["pnml_test_logloss"] < log_inf_limit]
#     res_df = res_df[res_df["mn_test_logloss"] < log_inf_limit]
#     res_df = res_df[res_df["trainset_size"] != res_df["num_features"]]
#     res_df = res_df[res_df["trainset_size"] != res_df["num_features"] + 1]
#     res_df = res_df[res_df["trainset_size"] != res_df["num_features"] - 1]

    print(
        f"num_featues={res_df.num_features.mean()}. trainset sizes={res_df.trainset_size.unique()}"
    )
    res_mean_df = res_df.groupby("trainset_size").mean().reset_index()
    res_std_df = res_df.groupby("trainset_size").agg(np.std, ddof=0).reset_index()
    res_count_df = res_df.groupby("trainset_size").agg("count")

    res_list.append(
        {
            "dataset_name": dataset_name,
            "mean_df": res_mean_df,
            "std_df": res_std_df,
            "count_df": res_count_df,
            "res_df": res_df.copy(),
        }
    )
    print()
```

```python
def plot_confidence_interval(ax, x, mean, std, count, color: str):
    mean, std, count = np.array(mean), np.array(std), np.array(count)
    lower = mean - 1.960 * std / np.sqrt(count)
    upper = mean + 1.960 * std / np.sqrt(count)
    ax.fill_between(x, upper, lower, color=color, alpha=0.2)  # std curves


def plot_logloss(ax, res_dict: dict):
    dataset_name = res_dict["dataset_name"]
    mean_df = res_dict["mean_df"]
    std_df = res_dict["std_df"]
    count_df = res_dict["count_df"]
    trainset_size = mean_df["trainset_size"].values

    key = "mn_test_logloss"
    ax.plot(trainset_size, mean_df[key], label="Minmum Norm", color="C0", alpha=0.6)
    plot_confidence_interval(
        ax, trainset_size, mean_df[key], std_df[key], count_df[key], "C0"
    )
    print(key, mean_df[key].max())

    key = "pnml_test_logloss"
    ax.plot(trainset_size, mean_df[key], label="pNML", color="C1", alpha=0.6)
    plot_confidence_interval(
        ax, trainset_size, mean_df[key], std_df[key], count_df[key], "C1"
    )
    print(key, mean_df[key].max())


#     key = "genie_test_logloss"
#     ax.plot(trainset_size, mean_df[key], label="Genie", color="C2")
#     plot_confidence_interval(
#         ax, trainset_size, mean_df[key], std_df[key], count_df[key], "C2"
#     )
    print(trainset_size)

def plot_regret(ax, res_dict: dict):
    mean_df = res_dict["mean_df"]
    std_df = res_dict["std_df"]
    count_df = res_dict["count_df"]
    trainset_size = mean_df["trainset_size"].values

    key = "pnml_regret"
    ax.plot(
        trainset_size, mean_df[key], label="Empirical", color=colors[3], alpha=0.6
    )
    plot_confidence_interval(
        ax, trainset_size, mean_df[key], std_df[key], count_df[key], colors[3],
    )

    key = "analytical_pnml_regret"
    ax.plot(
        trainset_size, mean_df[key], label="Analytical", color=colors[-4], alpha=0.6
    )
    plot_confidence_interval(
        ax, trainset_size, mean_df[key], std_df[key], count_df[key], colors[-4],
    )
```
